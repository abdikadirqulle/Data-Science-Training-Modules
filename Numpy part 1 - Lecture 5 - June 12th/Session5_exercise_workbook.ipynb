#### Exercise 1: Array Creation and Basic Operations
Task: Create two 1D NumPy arrays and perform basic arithmetic operations on them.

1. Import the necessary library.

2. Create two 1D arrays a and b with 10 elements each, using pre-filled arrays (np.arange and np.linspace).

3. Perform element-wise addition, subtraction, multiplication, and division of these arrays.

4. Compute the dot product of the arrays, i.e.: $a \cdot b$ = $\sum a_i \cdot b_i$

5. Print the results of these operations.

```python
import numpy as np

# Step 1: Import the necessary library (already done above)

# Step 2: Create two 1D arrays `a` and `b` with 10 elements each using pre-filled arrays
a = np.arange(1, 11)           # Array with values from 1 to 10
b = np.linspace(10, 1, 10)     # Array with 10 values evenly spaced from 10 to 1

# Step 3: Perform element-wise arithmetic operations
addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b

# Step 4: Compute the dot product of the arrays
dot_product = np.dot(a, b)

# Step 5: Print the results of these operations
print("Array a:", a)
print("Array b:", b)
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
print("Dot Product:", dot_product)
```

#### Exercise 2: Array Manipulation and Functions
Task: Use NumPy functions to manipulate arrays.

1. Create a 1D array x with 15 elements (you can use **np.arange** or **np.random.randint**)

2. Find the maximum, minimum, mean, and standard deviation of the array.

3. Create a new array y that contains the square of each element in x.
    - do this in 3 different ways

```python
import numpy as np

# Step 1: Create a 1D array `x` with 15 elements
x = np.arange(1, 16) # or x = np.random.randint(0, 100, size=15)

# Step 2: Find the maximum, minimum, mean, and standard deviation of the array
max_value = np.max(x)
min_value = np.min(x)
mean_value = np.mean(x)
std_deviation = np.std(x)

# Step 3: Create a new array `y` that contains the square of each element in `x`
y1 = np.square(x)
y2 = x*x
y3 = x**2

# Print the results
print("Original array:", x)
print("Max value:", max_value)
print("Min value:", min_value)
print("Mean value:", mean_value)
print("Standard Deviation:", std_deviation)
print("Squared array v1:", y1)
print("Squared array v2:", y2)
print("Squared array v3:", y3)
```

#### Exercise 3: Plotting with Matplotlib
Task: Create a plot using Matplotlib to visualize data.

1. Create a 1D array t representing time (from 0 to 2Ï€) with 100 elements.

2. Create two 1D arrays sin_t and cos_t that contain the sine and cosine of t, respectively.

3. Plot t vs. sin_t and t vs. cos_t on the same graph.

4. Add labels, a title, and a legend to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a 1D array `t` representing time (from 0 to 2Ï€) with 100 elements
t = np.linspace(0, 2 * np.pi, 100)

# Step 2: Create two 1D arrays `sin_t` and `cos_t` that contain the sine and cosine of `t`, respectively
sin_t = np.sin(t)
cos_t = np.cos(t)

# Step 3: Plot `t` vs. `sin_t` and `t` vs. `cos_t` on the same graph
plt.plot(t, sin_t, label='sin(t)')
plt.plot(t, cos_t, label='cos(t)')

# Step 4: Add labels, a title, and a legend to the plot
plt.xlabel('Time (t)')
plt.ylabel('Function values')
plt.title('Sine and Cosine Functions')
plt.legend()

# Display the plot
plt.show()
```

#### Bonus Exercise 4: Advanced Array Operations and Plotting
Task: Perform advanced array operations and visualize the results.

1. Import the necessary libraries.
2. Create a 1D array x with 100 elements, evenly spaced from -10 to 10.
3. Compute the following functions based on x:
    - $y_1 = x^2$
    - $y_2 = x^3$
    - $y_3 = sin(x)$

4. Create a new array combined which is the sum of y1, y2, and y3.
5. Plot x vs. y1, x vs. y2, x vs. y3, and x vs. combined on the same graph with appropriate labels and legend.

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Import the necessary libraries (already done above)

# Step 2: Create a 1D array `x` with 100 elements, evenly spaced from -10 to 10
x = np.linspace(-10, 10, 100)

# Step 3: Compute the functions based on `x`
y1 = x**2               # y1 = x^2
y2 = x**3               # y2 = x^3
y3 = np.sin(x)          # y3 = sin(x)

# Step 4: Create a new array `combined` which is the sum of `y1`, `y2`, and `y3`
combined = y1 + y2 + y3

# Step 5: Plot `x` vs. `y1`, `x` vs. `y2`, `x` vs. `y3`, and `x` vs. `combined` on the same graph
plt.plot(x, y1, label='$y1 = x^2$')
plt.plot(x, y2, label='$y2 = x^3$')
plt.plot(x, y3, label='$y3 = \sin(x)$')
plt.plot(x, combined, label='$combined = y1 + y2 + y3$', linestyle='--')

# Add labels, title, and legend to the plot
plt.xlabel('x')
plt.ylabel('Function values')
plt.title('Advanced Function Plots')
plt.legend()

# Display the plot
plt.show()
```
